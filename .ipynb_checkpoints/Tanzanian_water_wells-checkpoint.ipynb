# **Tanzanian Water Wells**

# Business Understanding

### Business Overview  

Tanzania is one of the countries located in East Africa and has a population of over 67 million. The country has a diverse geography that includes Mount Kilimanjaro, the Serengeti, and lakes Victoria and Tanganyika. Its home to over 120 ethnic groups and uses Swahili and English as official languages. Most people live in rural areas, and the population is largely young. Politically, it is a stable presidential republic led by the long-ruling Chama Cha Mapinduzi (CCM) party. The economy depends on agriculture, tourism, and mining. However, Tanzania faces key development challenges, including poverty, youth unemployment, and regional disparities. A critical concern is limited access to clean water, especially in rural areas where water sources are scarce or polluted. This impacts health, agriculture, and education, particularly for women and children who collect water daily.

With this context, the Government of Tanzania in collboration with sector stakeholders through the Ministry of Water and Irrigation is in progress implementing its 20-Year (2006-2025) Water Sector Development Programme (WSDP) alongside the Tanzania Development Vision 2025 which aspires to achieve universal access to water supply in urban areas and ensuring 90% water supply coverage in rural areas by 2025. The Vision also aspires to ensure that water resources are available in a sustainable manner to serve as a driver to social and economic needs. This goal aligns and contributes to the realization of Sustainable Development Goal 6, that aims to ensure the availability and sustainable management of water and sanitation for all by 2030. In Tanzania, the current average water supply coverage in urban and rural areas is 78% and 59.76%, respectively. The major water sources include Lake Victoria, Lake Tanganyika, rivers like the Rufiji, underground aquifers, and wells both shallow and deep. However, access is challenged by seasonal variability, pollution, overuse, and inadequate infrastructure. In rural areas, many depend on wells, which often face issues like contamination, drying, and poor maintenance.

### Problem Statement  

In Tanzania, a larger proportion of its citizenry relies of water wells as key source of water. Such sources are prone to contamination, drying, and poor maintenance. Therefore, it requires monitoring for water levels, quality and even maintenance. The Tanzanian water wells dataset provides a comprehensive information for approximately over 59,000 water wells with necessary features providing information related to pump, installation, location and condition related.  As government in collaboration with sector stakeholders (NGOs, private actors) put measures to achieve universal access to water supply in urban areas and ensuring 90% water supply coverage in rural areas by 2025. It is prudent to identify where to invest scarce resources and achieve optimal impact utilizing the available information. The task at hand involves finding a scientific approach that can provide information on which waterpoints (water wells) will likely fail and thus better plan accordingly for operational maintenance.
Therefore, as a data scientist, we hope to solve this problem by developing a predictive model that can be used to determine which water pumps are faulty to promote access to clean, potable water across Tanzania

#### Stakeholders
* The Tanzanian government and its Agencies (Ministry of Water and Irrigation)
* Non Governmental Organizations
* Private Actors

#### **Objectives**
The overall objective is to build a predictive model that classifies the operational status of water wells based on attributes provided in the Tanzanian Water Wells dataset. To realize this, the following specific objectives are introduced.  
1. To identify patterns influencing the functionality of the wells
2. Develop models for predicting the functionality of the water wells
3. To evaluate the models and identify the significant features for determining wells functionality 

#### Metrics for Model Evaluation 
To evaluate the performance of the models, the following metrics will be considered  
1. Accuracy - *Measure overall correctedness* 
2. Precision - *Measures the purity of the positive predictions: the proportion of items the model labelled as positive that are actually positive (TP / (TP + FP)*
3. Recall - *Measures the completeness of the positive class: the proportion of all true positives the model successfully recovered (TP / (TP + FN))*
4. Confusion matrix - *Visualize performance per class by showing true and predicted values*
5. ROC - AUC - *Evaluates ability to distinguish classes*

**Imputing missing values using the median is a common and practical strategy, especially for numeric features.**   
*Strengths*   
1.  Median is not affected by extreme values (unlike mean).This is helpful for skewed distributions like amount_tsh or population
2.  Prevents data loss by avoiding dropping rows with missing values.  

*Challenges*
1. Loss of Variance - Replacing many values with a the median flattens the distribution and may mask important patterns
2. Not Ideal for Linear Models such as logistic regression since many identical imputed values might reduce model sensitivity or introduce bias.

```python
# Impute only in the four numeric features of interest
cols_to_impute = ['amount_tsh', 'gps_height', 'population', 'construction_year']

# Replace NaN values in selected columns with their median
for col in cols_to_impute:
    median_val = water_wells_data[col].median()
    water_wells_data[col] = water_wells_data[col].fillna(median_val)
```

```python
df = pd.read_csv('Data/water_wells_data.csv')
target = "status_group"

#Numeric & target variable
num_cols = df.select_dtypes("float").columns

for col in num_cols:
    # one‑way ANOVA (≈ “Is the mean different across classes?”)
    groups = [df[df[target] == g][col] for g in df[target].unique()]
    F, p = ss.f_oneway(*groups)
    print(f"{col:<20}  ANOVA p‑value = {p:0.3g}")

# Box‑plots to visualise the spread
sns.boxplot(data=df.melt(id_vars=target, value_vars=num_cols),
            x=target, y="value", hue="variable")
plt.title("Numeric features by status group"); plt.show()

# Categorical & the target variable
cat_cols = df.select_dtypes("object").drop(columns=[target]).columns

def cramers_v(tbl):
    χ2 = ss.chi2_contingency(tbl, correction=False)[0]
    n   = tbl.sum().sum()
    r, k = tbl.shape
    return np.sqrt(χ2 / (n * (min(k, r) - 1)))

for col in cat_cols:
    ct = pd.crosstab(df[col], df[target])
    v  = cramers_v(ct)
    print(f"{col:<20}  Cramér’s V = {v:0.3f}")

#stacked bar chart
    (ct.div(ct.sum(1), 0)
       .plot(kind="bar", stacked=True, figsize=(4,2)))
    plt.title(col); plt.ylabel("row %"); plt.legend(bbox_to_anchor=(1,1))
    plt.show()
```

**Interpretation**  
* All four numeric predictors show statistically significant mean differences across the three status categories (ANOVA p < 10⁻⁶)
* The box‑plot reveals those differences are subtle for the bulk of wells and driven mainly by a long right‑tailed handful of very large values (especially for amount_tsh and population).
* Median GPS height and construction year appear slightly higher for functional wells, hinting that newer, higher‑elevation sites fare better, yet overlap remains substantial.
* For the first categorical check, extraction_type_class yields Cramér’s V ≈ 0.32, a moderate association meaning pump technology has some power to separate working from failed points but is far from deterministic.
* Every variable contributes some signal, but none alone will cleanly split the classes; combining them (and taming outliers) is essential for a performant model.

## Modelling

**Model 1 - Logistic Regression**  
A case of Multi class 

```python
#Load the clean dataset
water_wells_logreg = pd.read_csv('Data/water_wells_data.csv')
```

```python
water_wells_logreg.shape
```

```python
# Multinomial Logistic Regression modelling

# Define target and features
X = water_wells_logreg.drop('status_group', axis=1)
y = water_wells_logreg['status_group']
```

```python
# Identify categorical and numeric features
cat_cols = X.select_dtypes(include='object').columns.tolist()
num_cols = X.select_dtypes(include=['float64', 'int64']).columns.tolist()

preprocessor = ColumnTransformer([
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols),
    ('num', StandardScaler(), num_cols)
])
```

```python
# Logistic Regression with multinomial setting
model = Pipeline([
    ('prep', preprocessor),
    ('clf', LogisticRegression(solver='saga', max_iter=2000, multi_class='multinomial'))
])

# Split and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
model.fit(X_train, y_train)
```

```python
# Evaluate
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# Confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
pd.DataFrame(cm,
             index=pd.Index(model.classes_, name="Actual"),
             columns=pd.Index(model.classes_, name="Predicted"))
```

```python
#Confusion matrix - heat map plot
fig, ax = plt.subplots(figsize=(6, 6))
ConfusionMatrixDisplay.from_predictions(
    y_test,
    y_pred,
    display_labels=model.classes_,   # ['functional', 'functional needs repair', non-fuctional]
    cmap='Blues',                    
    xticks_rotation=45,
    normalize=None,
    ax=ax
)
ax.set_title("Multiclass Confusion Matrix")
plt.tight_layout()
plt.show()
```

**Model 1 Interpretation**   
Classification report & Confusion Matrix
* The *functional‑needs‑repair* class is invisible to the model, i.e it returns a Precision, recall and F‑score of 0.00. The model predicts every functional well that needs repair as being either functional (83 %) or non‑functional (17 %).  
* This can be attributable to class imbalance where in your training set only ≈ 5 % of wells are “functional and needs repair” or feature overlap due to limitations of linear models not being able to carve out a distinct decision region.
* The model finds almost all genuinely functional wells (few false‑negatives) but also over‑predicts “functional”, so ~31 % of its “functional” calls are wrong.
* The predictions of failure are usually correct (high precision) but it misses 36 % of truly failed wells (false‑negatives landing in the “functional” column).
* The logistic regression (model 1) learns to ignore the minority class to maximise overall accuracy.

**Model 2 - Logistic Model**  
A case of Binary Logistic Regresion

```python
water_wells_logreg2 = pd.read_csv('Data/water_wells_data.csv')
```

```python
# Replace 'functional needs repair' with 'non functional'
water_wells_logreg2['status_group'] = water_wells_logreg2['status_group'].replace('functional needs repair', 'non functional')
```

```python
print(water_wells_logreg2['status_group'].value_counts())
```

```python
water_wells_logreg2.info()
```

```python
# Define the target features
X = water_wells_logreg2.drop(columns=['status_group'])
y = (water_wells_logreg2['status_group'] == 'functional').astype(int)   # 1 = functional, 0 = non‑functional
```

```python
# Group the columns
cat_cols  = X.select_dtypes(include='object').columns.tolist()           
num_cols  = X.select_dtypes(include=['float64','int64']).columns.tolist()
```

```python
# Preprocessing and modelling
preprocess = ColumnTransformer([
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols),
        ('num', StandardScaler(),                        num_cols)
    ])
logreg = LogisticRegression(max_iter=2000, solver='lbfgs', class_weight='balanced')

from sklearn.pipeline import Pipeline
pipe = Pipeline([
        ('prep', preprocess),
        ('clf' , logreg)
    ])
```

```python
# Train and test split
X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, stratify=y, random_state=42)

pipe.fit(X_train, y_train)
```

```python
# Model Evaluation
y_pred = pipe.predict(X_test)
print(classification_report(y_test, y_pred))

y_prob = pipe.predict_proba(X_test)[:,1]
print("ROC‑AUC:", roc_auc_score(y_test, y_prob).round(3))
```

**Model 2 - Interpretation**   
Classification report  

* Non‑functional wells (class 0) have high precision (86 %) in that when the model says “broken” it is usually right. However, lower recall (59 %) implies it misses roughly 4 in 10 truly broken wells.

* Functional wells (class 1) also have high recall (90 %) implying that very few working wells are labelled broken. However, moderate precision (68 %) indicates that about one‑third of “functional” predictions are wrong. In business context, it relates to wasted technician visits if you act on them without inspection.

* Balanced class weights clearly aided recall on the minority class, but the linear boundary is still too simple to catch every pattern.

* The ROC curve score of 0.802 indicates that if the test set is shuffled and compared one positive vs. one negative at random, the model’s score is higher for the positive 80 % of the time. Therefore, Model 2 (balanced logistic-regression pipeline) is in the “good-but-not-yet-great” zone. It discriminates reasonably well between functional and non-functional wells, but there is room for improvement.

*Additional plots to support interpretation are documented below*

```python
# confusion_matrix where 0 = non‑functional, 1 = functional
cm = confusion_matrix(y_test, y_pred, labels=[0, 1])

pd.DataFrame(
    cm,
    index=pd.Index(['Actual 0 (non‑functional)', 'Actual 1 (functional)'], name='True label'),
    columns=pd.Index(['Pred 0', 'Pred 1'], name='Predicted')
)
```

```python
# Confusion matrix heat plot
fig, ax = plt.subplots(figsize=(6, 6))

ConfusionMatrixDisplay.from_predictions(
    y_test,
    y_pred,
    labels=[0, 1],
    display_labels=['non‑functional (0)', 'functional (1)'],
    cmap='Blues',
    values_format='d',     
    ax=ax                  
)

ax.set_title("Confusion Matrix – Balanced Logistic Regression")
plt.tight_layout()
plt.show()
```

```python
#Plot RoC Curve

fig, ax = plt.subplots(figsize=(5, 5))

RocCurveDisplay.from_predictions(
    y_test,
    y_prob,              # predicted probability for the positive class
    name=f"LogReg (AUC = {roc_auc_score(y_test, y_prob):.3f})",
    ax=ax
)

ax.plot([0, 1], [0, 1], linestyle="--")   # reference “no-skill” line
ax.set_xlabel("False-Positive Rate")
ax.set_ylabel("True-Positive Rate (Recall)")
ax.set_title("ROC Curve – Balanced Logistic Regression")
plt.tight_layout()
plt.show()
```

**Interpretation - RoC Curve**  
The ROC curve consistently stays relatively close to the top left corner, indicating the highest true positive rate and the lowest false positive rate across thresholds. Thus highest overall classification performance




**Alternative Models - Decision Tree**

**Model 3- Decision tree**  
(Binary response variable)

```python
water_wells_tree = pd.read_csv('Data/water_wells_data.csv')
```

```python
water_wells_tree['status_group'].value_counts()
```

```python
# Replace 'functional needs repair' with 'non functional'
water_wells_tree['status_group'] = water_wells_tree['status_group'].replace('functional needs repair', 'non functional')
```

```python
# Define the target variables
X = water_wells_tree.drop(columns="status_group")
y = (water_wells_tree["status_group"] == "functional").astype(int)   # 1 = functional, 0 = non‑functional
```

```python
# Pre-processing: One‑hot encode all object (categorical) columnsn while leave numeric columns unchanged (trees don’t need scaling)
cat_cols = X.select_dtypes(include="object").columns
num_cols = X.select_dtypes(exclude="object").columns

preprocess = ColumnTransformer([
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
        ("num", "passthrough",                             num_cols)
    ])
```

```python
# Modelling - Decision tree
tree_clf = DecisionTreeClassifier(
        criterion     = "gini",
        max_depth     = None,        
        min_samples_split = 2,
        class_weight  = "balanced", 
        random_state  = 42
    )

pipe = Pipeline([
        ("prep", preprocess),
        ("clf" , tree_clf)
    ])
```

```python
# Train and test split
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

pipe.fit(X_tr, y_tr)
```

```python
# Model Evaluation
y_pred  = pipe.predict(X_te)
y_prob  = pipe.predict_proba(X_te)[:,1]

print("Initial Decision Tree classification summary")
print(classification_report(y_te, y_pred))
print("ROC‑AUC:", roc_auc_score(y_te, y_prob).round(3))
```

**Hyper parameter tuning for Model 3**

```python
# hyper‑parameters search for the decision‑tree classifier 

pipe = Pipeline([
    ('prep', preprocess),
    ('clf', tree_clf)               # keep the original name
])

param_grid = {
    'clf__max_depth'       : [None, 10, 20, 40],
    'clf__min_samples_leaf': [1, 5, 10, 20]
}

grid = GridSearchCV(pipe, param_grid,
                    cv=5, scoring='f1_weighted', n_jobs=-1)
print("Tuned Decision Tree Classification Report")
grid.fit(X_tr, y_tr)
print(grid.best_params_)
print(classification_report(y_te, grid.predict(X_te)))
```

**Interpretation**
* The grid‑search tried four tree depths and four minimum leaf sizes, scoring each setting with 5‑fold weighted F1.
* The optimal configuration is of max_depth = 20, min_samples_leaf = 10 keeps the tree deep enough to capture feature interactions yet prunes very small leaves to curb over‑fitting.
* On the untouched test set it lifts overall accuracy to 0.76 and macro‑/weighted‑F1 to 0.76, edging past the earlier logistic model - Model 2.
* The recall for the class 0 (non‑functional wells) rises to 0.67 while maintaining high precision (0.83), meaning the tuned tree detects about 230 additional failures with only a modest increase in false alarms.
* Class 1 (functional) retains strong recall (0.86) but gains precision (0.72).
* The hyper‑parameter search has yielded a more balanced decision‑tree that better identifies broken water points without sacrificing too much reliability on the functional ones.

```python
#RoC curve plot for the tuned model
best_tree = grid.best_estimator_            # = pipeline with prep + clf

y_score = best_tree.predict_proba(X_te)[:, 1]   # probability of class 1

fig, ax = plt.subplots(figsize=(6, 6))

RocCurveDisplay.from_predictions(
    y_te,
    y_score,
    name="Tuned Decision Tree",
    ax=ax,
)

ax.plot([0, 1], [0, 1], linestyle="--", color="grey", lw=1)  # chance line
auc = roc_auc_score(y_te, y_score)
ax.set_title(f"ROC curve – Decision Tree (AUC = {auc:.3f})")

plt.tight_layout()
plt.show()
```

```python
# Confusion Matrix plot
y_pred = grid.predict(X_te)

fig, ax = plt.subplots(figsize=(6, 6))
ConfusionMatrixDisplay.from_predictions(
    y_te, y_pred,
    display_labels=['non‑functional (0)', 'functional (1)'],
    cmap='Blues', values_format='d', ax=ax
)
ax.set_title("Confusion Matrix – Tuned Decision Tree")
plt.tight_layout()
plt.show()
```

**Intepretation** - Tuned Model  
* The ROC curve shows how the tuned decision‑tree separates functional (positive) from non‑functional wells across every possible probability threshold. Since the blue line bows well above the diagonal, the model consistently delivers higher scores to true positives than to true negatives; the AUC of ≈ 0.82 quantifies this, meaning that in 82 % of random well‑pairs (one working, one broken) the model ranks the working well higher.
* The model can achieve around 80 % recall of functional wells while incurring only about 20 % false‑alarm rate.
* The curve’s smooth ascent toward the top‑left but slight sag in the middle signals good but not perfect discrimination

**Model 4 - Decision tree**  
Multi class

```python
water_wells_tree2 = pd.read_csv('Data/water_wells_data.csv')
```

```python
#Define the Target variables
X = water_wells_tree2.drop(columns="status_group")
y = water_wells_tree2["status_group"] # All the three categories kept
```

```python
#Pre-processing : One hot encoding for categorical and pass through for numerics
cat_cols = X.select_dtypes(include="object").columns     
num_cols = X.select_dtypes(exclude="object").columns       

preprocess = ColumnTransformer([
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
        ("num", "passthrough",                             num_cols)
    ])
```

```python
# Decision Tree modeling 
tree_clf = DecisionTreeClassifier(
        criterion          = "gini",
        max_depth          = None,          # tune later if desired
        min_samples_leaf   = 10,            # small value curbs over‑fitting
        class_weight       = "balanced",    # tackle class imbalance
        random_state       = 42
    )

# Put everything in a single pipeline
pipe = Pipeline([
        ("prep", preprocess),
        ("tree", tree_clf)
    ])
```

```python
# Train and test Split the data
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.20, stratify=y, random_state=42)

pipe.fit(X_tr, y_tr)
```

```python
# Model evaluation
y_pred = pipe.predict(X_te)
print(classification_report(y_te, y_pred))
```

**Interpretation** - Model 4  
* The decision‑tree (Model  4) correctly labels most truly *non‑functional* wells(67 % recall, 82 % precision) and finds nearly two‑thirds of the *fucntional needs‑repair* wells (63 % recall), but it does so very inefficiently with the vast majority of being flagged as needing repair while actually fine (only 12 % precision).
* However, he model is cautious about calling a well functional: when it does, it is right three‑quarters of the time (74 % precision) yet leaves almost half of the working wells mis‑classified (52 % recall).
* These trade‑offs pull overall accuracy down to 59 % and the macro‑averaged F1 to 0.51, revealing that performance is highly imbalanced across classes.
* If such a model was used in practice, many unnecessary maintenance dispatch can be done visits while still missing a substantial share of both functioning and non‑functioning pumps. This requires  further tuning to raise precision on the “needs‑repair” class without eroding its hard‑won recall.

**Tuned Model**

```python
tree_clf = DecisionTreeClassifier(class_weight='balanced', random_state=42)

pipe = Pipeline([
    ('prep', preprocess),   # your ColumnTransformer
    ('clf', tree_clf)
])

param_grid = {
    'clf__max_depth'       : [5, 10, 20, None],
    'clf__min_samples_leaf': [5, 10, 20],
    'clf__min_samples_split': [10, 20, 40],
    'clf__max_leaf_nodes'  : [None, 50, 100]
}

grid = GridSearchCV(
    pipe, param_grid,
    cv=5, scoring='f1_weighted', n_jobs=-1
)
grid.fit(X_tr, y_tr)
print("Best params:", grid.best_params_)
```

```python
y_pred = grid.predict(X_te); print(classification_report(y_te, y_pred))
```

```python
# Post‑pruning the tuned Decision‑Tree in a single run

#Define the base (pre‑tuned) tree 
base_tree = DecisionTreeClassifier(
    class_weight='balanced',
    random_state=42,
    max_depth=None,
    min_samples_leaf=5,
    min_samples_split=10
)

#Derive a sensible ccp_alpha grid from the cost‑complexity path
_ = preprocess.fit(X_tr)                           
X_tr_proc = preprocess.transform(X_tr)           

path = base_tree.cost_complexity_pruning_path(X_tr_proc, y_tr)
# Take ~20 evenly spaced, unique alphas between 0 and the max
ccp_grid = np.unique(np.linspace(0, path.ccp_alphas.max(), 20))

#Build a pipeline & run GridSearchCV
pipe_prune = Pipeline([
    ('prep', preprocess),           # your ColumnTransformer
    ('clf', base_tree)
])

param_grid_prune = {'clf__ccp_alpha': ccp_grid}

grid_prune = GridSearchCV(
    pipe_prune,
    param_grid=param_grid_prune,
    cv=5,
    scoring='f1_weighted',
    n_jobs=-1
)
print("Fitting pruned‑tree grid …")
grid_prune.fit(X_tr, y_tr)
print("Best ccp_alpha:", grid_prune.best_params_['clf__ccp_alpha'])
```

```python
# Evaluate on the test set 
y_pred = grid_prune.predict(X_te)
print("\nPruned tree – test‑set report\n",
      classification_report(y_te, y_pred))

fig, ax = plt.subplots(figsize=(5, 5))
ConfusionMatrixDisplay.from_predictions(
    y_te, y_pred,
    display_labels=['functional', 'needs repair', 'non‑functional'],
    cmap='Blues',
    values_format='d',
    ax=ax
)
ax.set_title("Confusion Matrix – Pruned Decision Tree")
plt.tight_layout()
plt.show()

# If you need the fitted, pruned model object:
pruned_model = grid_prune.best_estimator_
```

**Interpretation** - Model 4 - Pruned
* Pruning simplified the tree, boosting overall accuracy to 71 % and precision on non‑functional wells to 92 %, but recall for that class slipped to 51 %. The model now almost always labels *functional needs repair* wells as either functional or non-functional
* Recall and precision both zero, while identifying working wells with 97 % recall yet near middle line precision (64 %).
* Macro‑F1 falls to 0.48, revealing that the pruning cut variance at the cost of completely sacrificing the minority *functional needs repair* class.

**Model 5 - RandomForest**  
Multi-class

```python
water_wells_RF = pd.read_csv('Data/water_wells_data.csv')
```

```python
# Define target variables
X = water_wells_RF.drop(columns="status_group")
y = water_wells_RF["status_group"] 
```

```python
#Pre-processing : One hot encoding for categorical and pass through for numerics
cat_cols = X.select_dtypes(include="object").columns
num_cols = X.select_dtypes(exclude="object").columns

preprocess = ColumnTransformer([
    ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols), 
    ("num", "passthrough", num_cols)
])
```

```python
#Random Forest Modeling
rf_clf = RandomForestClassifier(
        n_estimators      = 300,
        max_depth         = None,
        min_samples_leaf  = 2,
        class_weight      = "balanced",
        n_jobs            = -1,
        random_state      = 42
    )

# Full pipeline
from sklearn.pipeline import Pipeline
pipe = Pipeline([
        ("prep", preprocess),
        ("rf"  , rf_clf)
    ])
```

```python
# Train and test split the data
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.20, stratify=y, random_state=42)

pipe.fit(X_tr, y_tr)
```

```python
#Model Evaluation
print("Random Forest classification")
y_pred = pipe.predict(X_te)
print(classification_report(y_te, y_pred))
```

```python
# Feature importance
ohe = pipe.named_steps["prep"].named_transformers_["cat"]
ohe_cols = ohe.get_feature_names_out(cat_cols)
all_cols = list(ohe_cols) + list(num_cols)

importances = pipe.named_steps["rf"].feature_importances_
fi = pd.Series(importances, index=all_cols).sort_values(ascending=False)
print("\nTop 15 feature importances:\n", fi.head(15))
```

**Model Evaluation - Comparisons**

#### Binary Models
Two models that consider a binary output variable (status group) were explored; That is **Model 2: Logistic regression** and **Model 3: Decision Tree**

*The tuned Model tree - Decision Tree outperforms the Model 2- Logistic model owing to the following factors*
* Accuracy rises from 0.74 to 0.76 and macro‑F1 from 0.74 to 0.76, thus mainly captures more non‑functional wells (recall 0.67 vs 0.59) while keeping high precision (0.83).

* Logistic regression still offers smoother probability ranking (ROC‑AUC 0.802) and simpler interpretation, but the tree’s richer splits deliver superior overall field usefulness in practice.

#### Multi-Class models
Three models were explored while keeping all categories of the status group variable; i.e the case of multi-calss predictive modelling. The models employed included **Model 1: Logistic regression**, **Model 4: Decision Tree** (including pruned) and **Model 5: Random Forest**  

* The Model 5 – Random Forest performs the best overall. It achieves the highest accuracy (0.67), macro average F1-score (0.63), and weighted average F1-score (0.71). It balances precision and recall well across all classes, especially handling the minority class "functional needs repair" better than others.
* Compared to logistic regression and decision trees, Random Forest shows improved generalization and classification performance, making it the most reliable model here.


#### Features importance
1. The most influential features in the Random Forest model are primarily numeric, with *gps_height*, *construction_year*, and *population* standing out as the top three.  
     * *gps_height* contributes the most, suggesting that elevation plays a key role in determining pump functionality of a water well, perhaps due to its influence on water pressure or accessibility.
     * *Construction_year* highlights the relevance of infrastructure age, implying older pumps are more likely to need repair.
     * The *population* emphasizes usage demand, with higher population may lead to faster wear or increased strain on the system.  

2. Also, categorical features  play a significant role, particularly in representing the condition, type, and usage of the water point.   
     * Features like *quantity_group_dry*, *extraction_type_class_other*, and *waterpoint_type_group_other* indicate that specific categories of water availability and extraction methods are strong signals of functionality of a water well.
     * Additionally, payment-related features such as *payment_type_never pay* and *payment_type_unknown* suggest that financial commitment and maintenance responsibility might influence whether a pump remains in working order.

3. Overall, the random forest model benefits from a diverse set of features both numeric and categorical which together provide a comprehensive picture of the factors affecting water pump performance.
     * The emphasis on both physical characteristics (like elevation and construction year) and socio-economic indicators (such as payment type and extraction method) reflects a balanced model that can make nuanced predictions. This insight can guide not only predictive modeling but also resource allocation for maintenance and infrastructure planning.
